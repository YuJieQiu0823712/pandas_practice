### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
disipline = euro_12[["Team","Yellow Cards",'Red Cards']]
disipline.head()
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
disipline.sort_values(["Red Cards","Yellow Cards"], ascending=False)
```

```python
round(disipline["Yellow Cards"].mean()) #
```
